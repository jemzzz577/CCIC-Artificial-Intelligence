## 📝 My Question
Based on a sample of 996 Spotify songs, what is the mean song duration?

## 📊 My Study


## 🔬 My Analysis - i wonder if I can do this whole thing in one cell lol

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('random_songs (1).csv')

duration_col = None
for col in df.columns:
    if 'duration' in col.lower() or 'length' in col.lower() or 'time' in col.lower():
        duration_col = col
        break

if duration_col is None:
    raise ValueError("Could not find a duration column. Check your CSV columns.")

if df[duration_col].max() > 1000:
    df['duration_sec'] = df[duration_col] / 1000
else:
    df['duration_sec'] = df[duration_col]

mean_duration = df['duration_sec'].mean()
std_duration = df['duration_sec'].std()

print(f"Duration column detected: {duration_col}")
print(f"Mean Duration: {mean_duration:.2f} seconds")
print(f"Standard Deviation: {std_duration:.2f} seconds")

plt.hist(df['duration_sec'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Song Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.show()

n = len(df['duration_sec'])
sem = std_duration / np.sqrt(n)
ci = stats.t.interval(0.95, df=n-1, loc=mean_duration, scale=sem)
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f}) seconds")

prob_over_4min = (df['duration_sec'] > 240).mean()
print(f"Probability a song is longer than 4 minutes: {prob_over_4min:.2%}")
```

## ✅ My Answer

Based on a sample of 996 Spotify songs, the average song duration is around 201.08 seconds (about 3 minutes and 21 seconds), and the standard deviation is about 161.00 seconds. This means that while most songs are around this average length, there is a wide variation, with some songs being much shorter or longer. The 95% confidence interval for the mean duration is between 191.07 and 211.09 seconds, showing that we can be reasonably confident that the true average duration of all Spotify songs falls within this range. Also, about 21.69% of songs are longer than 4 minutes, showing that longer songs are relatively uncommon but not rare.
