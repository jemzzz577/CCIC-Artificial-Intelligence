# In Monopoly, given that you land on a blue space, what is the probability that it's Boardwalk?

The sample space is Boardwalk.

I am measuring the probability of landing on Boardwalk.

```python
import seaborn as sns
import random
```

```python
blue_spaces = ["Oriental Ave", "Vermont Ave", "Connect Ave", "Park Place", "Boardwalk"]

trials = 100
results = [random.choice(blue_spaces) for _ in range(trials)]

probability_sim = results.count("Boardwalk") / trials
print("Simulated probability of Boardwalk:", probability_sim)

probs = [results.count(s)/trials for s in blue_spaces]

sns.barplot(x=blue_spaces, y=probs, palette="Blues")
```
