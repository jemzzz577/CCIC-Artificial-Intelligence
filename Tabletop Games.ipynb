# In Tic-Tac-Toe, what is the probability X wins each game?

The type of study I plan to conduct is an observational study. I am going to be observing the data that is produced by the simulation of the game. The outcome I am measuring is if X will win the game. What combinations will appear during the game? There is only two teams, and 9 available spaces before the game starts.

```python
import random
import seaborn
```

The first symbol to get three in a row is the winner. How will I do this?

```python
import random

print(random.choice(["X", "O"]))
```

I will use a "for loop" to print the symbols 9 times given how many slots are in the game normally.

```python
import random
for trial in range(100):
    print(f"Board {trial + 1}:")
    for row in range(3):
        for col in range(3):
            print(random.choice(["X", "O"]), end=" ")
        print()
    print("-" * 10)
```

```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x_count = 0
o_count = 0

for _ in range(100):
    for _ in range(9):
        mark = random.choice(["X", "O"])
        if mark == "X":
            x_count += 1
        else:
            o_count += 1

data = pd.DataFrame({
    'Symbol': ['X', 'O'],
    'Count': [x_count, o_count]
})

sns.barplot(data=data, x='Symbol', y='Count', palette='pastel')
plt.title('Total X and O Counts in 100 Random Boards')
plt.ylabel('Count')
plt.show()
```

```python
total_slots = 100 * 9

prob_x = x_count / total_slots

print(f"Probability of X: {prob_x:.4f} ({prob_x*100:.2f}%)")
```
