## Movies and TV Part 3
> What is the probability that I select a movie that has a popularity score over 7.0?

## Simulation

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt

np.random.seed(42)
popularity_scores = np.random.normal(loc=6.3, scale=1.2, size=1000)
df = pd.DataFrame({"Popularity": popularity_scores})
empirical_prob = (df["Popularity"] > 7.0).mean()
mean = df["Popularity"].mean()
std = df["Popularity"].std()
theoretical_prob = 1 - norm.cdf(7.0, loc=mean, scale=std)

print(f"Sample Mean: {mean:.2f}")
print(f"Sample Std Dev: {std:.2f}")
print(f"Empirical Probability (Popularity > 7.0): {empirical_prob:.3f}")
print(f"Theoretical Probability (Popularity > 7.0): {theoretical_prob:.3f}")

data = pd.DataFrame({
    "Type":["Empirical", "Theoretical"],
    "Probability": [empirical_prob, theoretical_prob]
})

sns.histplot(df["Popularity"], bins=30, kde=True, color="skyblue")

plt.axvline(7.0, color="red", linestyle="--", label="Threshold = 7.0")
plt.fill_betweenx(
    y=[0, plt.gca().get_ylim()[1]],
    x1=7.0,
    x2=plt.gca().get_xlim()[1],
    color="red",
    alpha=0.2,
    label="Popularity > 7.0"
)

plt.title("Distribution of Movie Popularity Scores")
plt.xlabel("Popularity Score")
plt.ylabel("Frequency")

plt.text(5.0, 50, f"Empirical: {empirical_prob:.3f}", fontsize=10, color="black")
plt.text(5.0, 40, f"Theoretical: {theoretical_prob:.3f}", fontsize=10, color="black")

plt.legend()
plt.show()
```

## Conclusion
> The empirical probability of choosing 7.0 is about 0.280% and the theoretical probability is 0.282%. The mean of this sample was 6.32 and the standard deviation was 1.18.
